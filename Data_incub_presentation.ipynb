```python
df = df.drop(['TRANSACTION_CODE','UNIT','ACTION_INDICATOR','ORDER_FORM_NO','CORRECTION_NO','STRENGTH','Measure','MME_Conversion_Factor'], axis=1)
df
```

```python
df['TRANSACTION_DATE'] = df['TRANSACTION_DATE'].apply(lambda x: pd.to_datetime(str(x), format='%m%d%Y'))
df['TRANSACTION_DATE']
```
